### Image Pyramids

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('data/butterfly.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20, 20))

plt.subplot(2, 2, 1)
plt.title("Original")
plt.imshow(image)

# down scales the original image by half (original_image/2)
smaller = cv2.pyrDown(image)

# upscales the image twice (original_image * 2)
larger = cv2.pyrUp(smaller)

plt.subplot(2, 2, 2)
plt.title("Smaller")
plt.imshow(smaller)

plt.subplot(2, 2, 3)
plt.title("Larger")
plt.imshow(larger)
```
