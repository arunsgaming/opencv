### 3.Dilation, Erosion, Opening and Closing

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('data/LinuxLogo.jpg')
image = cv2.cvtColor(image, 0)

plt.figure(figsize=(20, 20))
plt.subplot(3, 2, 1)
plt.title("Original")
plt.imshow(image)

# Let's define our kernel size
kernel = np.ones((3,3), np.uint8)

# Now we erode
erosion = cv2.erode(image, kernel, iterations = 1)

plt.subplot(3, 2, 3)
plt.title("Erosion")
plt.imshow(erosion)


# Let's define our bigger kernel size
kernel = np.ones((7,7), np.uint8)

# Now we erode
erosion = cv2.erode(image, kernel, iterations = 1)

plt.subplot(3, 2, 4)
plt.title("Erosion (increased kernel)")
plt.imshow(erosion)


# dilation with kernel 1 (increases white area)
kernel = np.ones((3,3), np.uint8)
dilation = cv2.dilate(image, kernel, iterations = 1)
plt.subplot(3, 2, 5)
plt.title("Dilation")
plt.imshow(dilation)

# dilation with increased iteration
dilation = cv2.dilate(image, kernel, iterations = 2)
plt.subplot(3, 2, 6)
plt.title("Dilation")
plt.imshow(dilation)

```

### morphologyEx

```python
image = cv2.imread('data/noisy_charliechaplin.png')
image = cv2.cvtColor(image, 0)

plt.figure(figsize=(30, 30))
plt.subplot(3, 2, 1)
plt.title("Original")
plt.imshow(image)

kernel = np.ones((3,3), np.uint8)

# Opening - Good for removing noise
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
plt.subplot(3, 2, 3)
plt.title("Opening")
plt.imshow(opening)


# Closing - Good for removing noise
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
plt.subplot(3, 2, 4)
plt.title("Closing")
plt.imshow(closing)
```
